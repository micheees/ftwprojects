## Changing Fam Income Class

```python
df.head()
```

```python
df2 = df.loc[df['Status'].isin(['Graduated', 'Disqualified'])] #Excluding Ongoing
df2.rename(columns = {' Annual Family Income ':'annual_fam_income', "Father's Annual Salary" :'f_income', "Mother's Annual Salary": 'm_income'}, inplace = True) #Renaming Columns
```

```python
df2.info()
```

```python
df_inc = df2[['Scholar ID', 'Status', 'annual_fam_income', 'f_income', 'm_income']] #Selecting Relevant Columns

df2['annual_fam_income'] = df2['annual_fam_income'].str.replace(',','') #Replacing Strings
df2['f_income'] = df2['f_income'].str.replace(',','')
df2['m_income'] = df2['m_income'].str.replace(',','')

df_inc['annual_fam_income'] = df2['annual_fam_income'].astype(float)
df_inc['f_income'] = df2['f_income'].astype(float)
df_inc['m_income'] = df2['m_income'].astype(float)

df_inc.head()
```

```python
df_inc.info()
```

```python
ax = sns.FacetGrid(df_inc, hue="Status", size=4, aspect=2, palette = 'rocket',).map(plt.hist, "annual_fam_income", histtype='barstacked', alpha = 0.5, stacked=True)
plt.legend(labels=['Graduated','Disqualified'])
```

```python
#https://pidswebs.pids.gov.ph/CDN/PUBLICATIONS/pidspn1818.pdf - Source for 2017 prices 

poor = 114240
low= 228480
lowmid = 456960
midmid = 799680
uppmid = 1370880
upper = 2284800
rich = 3600000

upd_limit2014 = 500000
upd_limit20082012 = 300000
provincial_limit = 250000
```

```python
df_inc.head()
```

```python
def convert_income(x):
    income = x[0]

    if income <= poor:
      return 'Poor'
    elif income <= low:
      return 'Lower'
    elif income <= lowmid:
      return 'Lower Middle'
    elif income <= midmid:
      return 'Middle Middle'  
    elif income <= uppmid:
      return 'Upper Middle'
    elif income <= upper:
      return 'Upper'            
    else:
        return 'Rich'

df_inc['class'] = df_inc[['annual_fam_income']].apply(convert_income,axis=1)
df_inc.head()
```

```python
sns.catplot(x='class', col = 'Status',data=df_inc, kind='count', palette='rocket')
```
